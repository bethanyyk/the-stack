```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Responsive Records - Batch II (24-3054).csv')
df.head()
```

```python
# Q2

averages = {}
counts = {}
order = [1, 2, 3, 4, 5, np.nan]
plots = {}

scale = {
    'Very dissatisfied': 1,
    'Dissatisfied': 2,
    'Neutral': 3,
    'Satisfied': 4,
    'Very Satisfied': 5
}

for col in ['Q2_1', 'Q2_2', 'Q2_3', 'Q2_4', 'Q2_5', 'Q2_6']:
    df[col] = df[col].map(scale)
    
    avg = df[col].mean()
    averages[col] = avg
    
    count = df[col].value_counts(dropna=False).reindex(order)
    counts[col] = count

    plt.figure(figsize=(4, 2))
    ax = sns.barplot(x=count.index, y=count.values)
    plots[col] = plt.gcf()
```
